**<h2>Data Overview</h2>**

<p>This project uses the two datasets <code>players.csv</code> and <code>sessions.csv</code> which contain information about individual player's and their play sessions.</p> 

**<code>players.csv</code>**
- **Number of observations:** 196 
- **Number of variables:** 7
- **Description:** Contains general background information on the player
- **Purpose:** Provides statistics like experience, age, gender, hours played that may help explain engagement or performance patterns.

| Variable Name  | Type      | Description                                                 |
| -------------- | --------- | ----------------------------------------------------------- |
| `hashedEmail`  | Character | Unique identifier for each player                           |
| `name`         | Character | Player name                                                 |
| `gender`       | Character | Self-reported gender (7 levels)                             |
| `Age`          | Double    | Age in years                                                |
| `experience`   | Character | Player experience level (e.g., Amateur, Intermediate, etc.) |
| `subscribe`    | Boolean   | Whether the player subscribes to a game-related newsletter  |
| `played_hours` | Double    | Total hours played                                          |




**<code>sessions.csv</code>**
- **Number of observations:** 1,535
- **Number of variables:** 5
- **Description:** Each row represents a single gameplay session
- **Purpose:** Tracks when sessions occured 

| Variable Name         | Type      | Description                           |
| --------------------- | --------- | ------------------------------------- |
| `hashedEmail`         | Character | Unique ID linking sessions to players |
| `start_time`          | Character | Session start time                    |
| `end_time`            | Character | Session end time                      |
| `original_start_time` | Double    | Unix timestamp for start              |
| `original_end_time`   | Double    | Unix timestamp for end                |

**Potential Issues in Data**
- in ```players.csv```: ```Age``` has 2 missing values
- in ```sessions.csv```: end_time has 2 missing values
- in ```players.csv```: ```played_hours``` is skewed with most values near 0
- in ```sessions.csv```: ```start_time``` and ```end_time``` are stored as characters, not a date object
- in ```players.csv``` and ```sessions.csv```: looking at ```hashedEmail``` (present in both datasets), we can see that only 125/196 players have session records.
- not all players have recorded sessions (sampling bias) 
- self-reported ages and genders may not be accurate
- timestamps may not be accurate (inconsistent timezones, system clock errors)



**Data Collection**
- Data appears to be collected via an automated software which tracks player data and session data.
- ```players.csv``` most likely comes from user registration / profile. 
- According to the research group's website, gameplay, speech, and key presses are all recorded.
- The timeframe appears to be from June-July 2024.
- Emails have been hidden via hashing.
