**<h2>Question</h2>**

**What player characteristics and behaviours are most predictive of subscribing to a game-related newsletter, and how do these features differ between various player types?**

In terms of predictor variables: **Can player stats like age, experience level, gender, and total hours played predict whether a player subscribes to a game-related newsletter in the ```players.csv``` dataset?**

The ```players.csv``` dataset provides player information that is directly relevant to predicting newsletter subscription. The response variable, ``subscribe,`` indicates whether the player subscribes to a game-related newsletter. Several explanatory variables are available to help explain this outcome, including demographic features such as ``Age`` and ``gender``, as well as behavioural and experience-based variables such as ```experience``` (e.g., Amateur, Intermediate, Expert) and ```played_hours```.

To address the question, we can first tidy and prepare the data by ensuring that categorical variables (such as ```gender``` and ```experience```) are properly encoded as factors, and by checking for missing values or outliers in numeric variables like ```Age``` and ```played_hours```. Once the dataset is tidy, I can apply a K-NN model to predict the response variable ```subscribe``` based on the predictor variables mentioned above. 

Note that the ```name``` and ```hashedEmail``` variables are not being considered due to them most likely not being relevant in this analysis. The visualizations below are done using the unchanged ```players.csv``` dataset, as it is already tidy enough for our current needs, however some changes were made while plotting (such as scaling some axes).

````r
players_means <- players |> 
    summarize(mean_age = round(mean(Age, na.rm = TRUE), 2), mean_played_hours = round(mean(played_hours, na.rm = TRUE), 2))

players_means
````

The average player is approximately **21** years old and has played for an average of **5.85** total hours.

**<h1>Exploratory Visualizations</h1>**

````r
ggplot(players, aes(x = Age)) +
  geom_histogram(binwidth = 2, fill = 'steelblue', color = 'white') +
  labs(
    title = 'Distribution of Player Ages',
    x = 'Age (years)',
    y = 'Number of Players'
  )
````

Most players are in their late teens to mid-20s, with fewer older players. This suggests that the player base skews toward younger audiences.

````r
ggplot(players, aes(x = subscribe, y = log10(played_hours + 1), fill = subscribe)) + # we scaled played_hours since it is highly skewed, scaling will reduce impact of large outliers
  geom_boxplot() +
  labs(
    title = 'Played Hours by Subscription Status',
    x = 'Subscribed to Newsletter',
    y = 'Log10 of Total Hours Played'
  ) + theme(legend.position = 'none')
````

Players who have subscribed tend to have higher played hours on average, though there is high variability. This suggests that ```played_hours``` could be a predictor of ```subscribe```.

````r
ggplot(players, aes(x = experience, fill = subscribe)) +
  geom_bar(position = "fill") +
  labs(
    title = "Subscription Rate by Experience Level",
    x = "Experience Level",
    y = "Proportion of Players"
  ) +
  scale_y_continuous(labels = scales::percent)
````

More experienced players (e.g., expert or intermediate) show higher proportions of being subscribed compared to beginners or amateurs. This could indicate that players who are more engaged or skilled are more likely to subscribe to newsletters

````r
ggplot(players, aes(x = gender, fill = subscribe)) +
  geom_bar(position = "fill") +
  labs(
    title = "Subscription Rate by Gender",
    x = "Gender",
    y = "Proportion of Players"
  ) +
  scale_y_continuous(labels = scales::percent)
````

The gender distribution is somewhat imbalanced (majority male), but the proportion of subscribers appears fairly similar across genders. This suggests gender may not be a strong predictor of subscription on its own.

In general: 
- ```age``` is mostly around 20 
- ```played_hours``` is highly skewed
- subscribed players tend to have more hours played and higher skill level 
- gender is likely not influential compared to other variables 
- given these relationships, we can use a classification model later on

**<h1>Methods and Plan</h1>**

- **Intended method:** K-NN classification to predict ```subscribe``` based on ```Age```, ```experience```, ```played_hours```.
- **Why this method is appropriate:** K-NN is suitable because ```subscribe``` is a binary variable, and K-NN can classify players by comparing them to others with similar characteristics. It does not assume a specific relationship between predictors and the outcome, making it flexible for this dataset.
- **Assumptions required:** The predictors must be scaled so that variables measured on different ranges contribute equally to distance calculations.
- **Potential limitations or weaknesses:** K-NN can be sensitive to irrelevant outliers.
- **Model comparison and selection:** 5-fold cross-validation will be used on the training data to test multiple k values. The value of k that minimizes error or maximizes accuracy will be selected as the final model.
- **Data processing:** Predictors will be standardized, categorical variables will be encoded as factors, data will be split into a training set (75%) and a testing set (25%) using a fixed random seed. Cross-validation will be performed on the training set, and the finalized model will be evaluated on the test set for accuracy.
